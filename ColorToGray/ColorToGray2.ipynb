```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
Src=Image.open('image.png')
SrcMatrix=np.array(Src)
```

```python
plt.imshow(SrcMatrix[:,:,:])
```

```python
print('Image dimensions:', SrcMatrix.shape)
```

```python
def SameWeight(image):
    return np.dot(image,[0.33,0.33,0.33])
same_weight = SameWeight(SrcMatrix)
plt.imshow(same_weight, cmap='gray')
```

```python
def Max(image):
    return np.max(image,axis=2)
max =Max(SrcMatrix)
plt.imshow(max, cmap='gray')
```

```python
def Min(image):
    return np.min(image,axis=2)
min =Min(SrcMatrix)
plt.imshow(min, cmap='gray')
```

```python
def Red(image):
    return image[:,:,0]
red =Red(SrcMatrix)
plt.imshow(red, cmap='gray')
```

```python
def Green(image):
    return image[:,:,1]
green=Green(SrcMatrix)
plt.imshow(green, cmap='gray')
```

```python
def Blue(image):
    return image[:,:,2]
blue=Blue(SrcMatrix)
plt.imshow(blue, cmap='gray')
plt.show()
```
